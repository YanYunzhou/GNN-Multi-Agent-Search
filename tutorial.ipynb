```python
from expert.SIPPPlanner_Naive_advance3 import SIPP_PP
path_dict, agents_time_list = SIPP_PP(env, env.agent_starts_vidxs, env.agent_goals_vidxs)
max_timestep=0
for i in path_dict.keys():
    current_path=path_dict[i]
    current_time_list=agents_time_list[i]
    current_path_len=len(current_path)
    for j in range(current_path_len-1):
        if current_path[j]!=current_path[j+1]:
            if current_time_list[j+1]-current_time_list[j]>max_timestep:
                max_timestep=current_time_list[j+1]-current_time_list[j]
print(max_timestep)
```
